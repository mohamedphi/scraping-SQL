```python
import requests
from bs4 import BeautifulSoup
import sqlite3
```

```python
connection=sqlite3.connect('books_toscrape.db')
```

# books of travel

```python
response=requests.get("https://books.toscrape.com/catalogue/category/books/travel_2/index.html")
soup=BeautifulSoup(response.text,"html.parser")

#Names of books
names=[]
for i in range(len(soup.find_all("h3"))):
    names_books=soup.find_all("h3")[i].find("a").get("title")
    names.append(names_books)


#Rating of books
rating=[]
for i in range(len(soup.find_all("h3"))):
    rating_books=soup.find_all("article")[i].find("p").get("class")[1]
    rating.append(rating_books)

#price of books
price=[]
for i in range(len(soup.find_all("h3"))):
    price_books=soup.find_all("div",attrs={"class":"product_price"})[i].find("p").text.split("£")[1]
    price.append(price_books)


y=list(zip(names,rating,price))


for i,x,z in y:
    cursor=connection.execute('insert into travel (Names,Rating,Price) values (?,?,?)',[i,x,z])
    connection.commit()
```
